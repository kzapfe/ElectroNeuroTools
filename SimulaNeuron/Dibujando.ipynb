```julia
using PyPlot, DelimitedFiles, JLD
```

```julia

lfpch=readdlm("lfpsimuladotalcual01.tsv")
```

```julia

lfp3=reshape(lfpch,801,33,29);
```

```julia
lfp3*=1e6; #microVolts
```

```julia
celula=read("cell.dat",String);
```

```julia
cel2=readdlm.(IOBuffer.(split(strip(celula),"\n\n")));
```

```julia
PyPlot.ioff()
f=figure(figsize=(4,4))
imshow(lfp3[400,:,:], extent=[-700,700,-400,1200], origin="lower")
for l in cel2
    plot(l[:,1], l[:,2], marker="None", linewidth=1)
end
#savefig("tuabuela.png",dpi=72)
#close()
```

```julia
vbass,valt=extrema(lfp3)./5
```

```julia
#= PyPlot.ioff()

f=figure(figsize=(4,3))
for l in cel2
    rayas=plot(l[:,1], l[:,2], marker="None", linewidth=1, zorder=4)
end
   imagen=imshow(lfp3[400,:,:], extent=[-700,700,-400,1200], 
       origin="lower", zorder=2, vmin=vbass, vmax=valt,
        cmap="plasma")
    cbr=colorbar(imagen)

for t=1:801
    imagen=imshow(lfp3[t,:,:], extent=[-700,700,-400,1200], 
        origin="lower", zorder=2, vmin=vbass, vmax=valt,
        cmap="plasma")
    nstring=lpad(t, 4, "0")
    nom="otro-$nstring"*".png"
    savefig(nom,dpi=72)
    #close()
    imagen.remove()
    if mod(t,50)==0
        println("vamos en $t")
    end
end
=#
```

```julia
csdarx=load("csdaejemplo01.jld")
csda=csdarx["csda"];
cmin,cmax=extrema(csda)
```

```julia
limin=-2.0 #cmin/100
limax=abs(limin)
```

```julia
f=figure(figsize=(4,3))
for l in cel2
    rayas=plot(l[:,1], l[:,2], marker="None", linewidth=0.2, zorder=4, color="grey")
end
   imagen=imshow(csda[400,:,:], extent=[-700,700,-400,1200], 
       origin="lower", zorder=2, vmin=limin, vmax=limax,
        cmap="bwr")
    cbr=colorbar(imagen)

for t=1:801
    imagen=imshow(csda[t,:,:], extent=[-700,700,-400,1200], 
        origin="lower", zorder=2, vmin=limin, vmax=limax,
        cmap="bwr")
    nstring=lpad(t, 4, "0")
    nom="csd-$nstring"*".png"
    savefig(nom,dpi=72)
    #close()
    imagen.remove()
    if mod(t,50)==0
        println("vamos en $t")
    end
end
```
