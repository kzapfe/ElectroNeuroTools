```julia
using MAT 
DatosAparato=matopen("propagacion.mat");
matread("propagacion.mat")
```

```julia
LasMedicionesCrudas=read(DatosAparato, "RawMatrix");
size(LasMedicionesCrudas)
```

```julia
ParteInteresante=LasMedicionesCrudas[:,:,1300:2700];
medidasdatos=size(ParteInteresante)
tmax=medidasdatos[3]
```

```julia
Example=reshape(ParteInteresante[50,50,:],tmax);
```

```julia
FFTExample=fft(Example);
tururu=real(FFTExample);
```

```julia
writedlm("Exemple-50-50.dat", Example)
writedlm("ExempleFFTW-50-50.dat", tururu)
```

```julia
for t=200:tmax
    FFTExample[t]=0
end
Exdenuez=ifft(FFTExample);
tarara=real(Exdenuez)*14.0/2.0;
```

```julia
using PyPlot
```

```julia
plot(Example[400:600], "r-", linewidth=2 , tarara[400:600], "g-", linewidth=0.5)
```
