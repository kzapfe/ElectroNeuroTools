Ahora toca hacer un supuesto "smoothing" de los datos. Se me ocurre lo siguiente: promediar los datos de 50 en 50 frames y ver que pasa, moviendo la ventana, claro esta, paso por paso.

```julia
function SuavizarDatos(data, param)
    tantos=length(data)
    result=zeros(data)
    for j=1:(tantos-param-1)
        chusco=sum(data[j:j+param])/param
        result[j]=chusco
    end
    return result
end
```

```julia
ChorizoExemplo=reshape(LasMedicionesCrudas[30,22,:],4213);
```

```julia
Hola=SuavizarDatos(ChorizoExemplo,50);
```

```julia
writedlm("data001.dat", Hola )
```

```julia
writedlm("datasinsmoot000.dat", ChorizoExemplo)
```

```julia
function desviaporventanas(data, param)
    tantos=length(data)
    result=zeros(data)
    for j=1:(tantos-param-1)
        chusco=std(data[j:j+param])
        result[j]=chusco
    end
    return result
end
    
```

```julia
goragora=desviaporventanas(ChorizoExemplo, 50);
writedlm("datadesviacion050.dat", goragora)
```
