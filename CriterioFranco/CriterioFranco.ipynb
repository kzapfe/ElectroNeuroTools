```julia
function SigmaVentanas(Datos, ancho, paso)
    anchomedio=floor(ancho/2)
    tantos=convert(Int, floor((length(Datos)-ancho)/paso))
    result=zeros(tantos)
    for t=1:tantos
        result[t]=std(Datos[(t-1)*paso+1:(t-1)*paso+ancho])
    end
    return result  
end
```

Este es un comentario con texto en latex $V=  \int dx f(x)$

```julia
ancho=350
paso=35
cuantasventanas=convert(Int, floor((tmax-ancho)/paso))
DesviacionPorVentana=zeros((64,64,cuantasventanas))
for j=1:64, k=1:64
    chorizo=reshape(ParteInteresante[j,k,:],tmax)
    DesviacionPorVentana[j,k,:]=SigmaVentanas(chorizo, ancho, paso)
end
```

```julia
#Estas son las medias y las desviaciones de LAS DESVIACIONES VENTANEADAS, no de los datos
mediasenventana=zeros(cuantasventanas)
desviacionesenventana=zeros(cuantasventanas)
for t=1:cuantasventanas
    mediasenventana[t]=mean(DesviacionPorVentana[:,:,t])
    desviacionesenventana[t]=std(DesviacionPorVentana[:,:,t])
end
```
