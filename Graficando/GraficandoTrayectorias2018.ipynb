# Graficas de Trayectorias para Evocados
Vamos a sacar unas buenas graficas de trayectorias evocadas

```julia
using JLD, PyPlot, HDF5
```

```julia
nombregeneral="../JLDFiles/TrabajandoCentrosdeMasa/140116s2cut2_evento_2"
nombrecsd=string(nombregeneral, ".jld")
nombretray=string(nombregeneral, "-Tray.jld")
```

```julia
datoscsd=load(nombrecsd)
datostray=load(nombretray)
```

```julia
csd=datoscsd["CSDLindenberg"]
#solo si es evocado hay retrazo claro
#retrazo=datoscsd["retratrazo"]
retrazo=1000
freq=datoscsd["freq"]
lfp=datoscsd["LFPTotal"]
saturados=datoscsd["CanalesSaturados"]
respuesta=datoscsd["Canalesrespuesta"]
    
cadpos=datostray["CatenarioPositivo"]
cadneg=datostray["CatenarioNegativo"];
```

```julia
figura=figure(figsize=(4,4))
xlim(0,65)
ylim(0,65)
limites=20
t=1200
ejemplo=csd[:,:,t]
tiempo=round((t-retrazo)/freq)

tick_params(
    axis="both",          # changes apply to the x-axis
    which="both",      # both major and minor ticks are affected
   # bottom="off",      # ticks along the bottom edge are off
    top="off",
    right="off",
   # left="off",
# ticks along the top edge are off
   # labelbottom="off",
    labelleft="on")

    title("t= $tiempo ms")
    imagen=imshow((ejemplo), origin="lower", interpolation="none",
    cmap="bwr", 
                      vmin=-limites,vmax=limites, extent=[0,64.5,0.5,64.5])

cbx=colorbar(imagen, fraction=0.045, ticks=[]) 

cbx[:set_label]("<-Sink | Source->")
```

```julia
figura=figure(figsize=(5,5))
xlim(0,65)
ylim(0,65)
        
limites=20
t=1200
ejemplo=csd[:,:,t]
tiempo=round((t-retrazo)/freq,1)

tick_params(
    axis="both",          # changes apply to the x-axis
    which="both",      # both major and minor ticks are affected
   # bottom="off",      # ticks along the bottom edge are off
    top="off",
    right="off",
   # left="off",
# ticks along the top edge are off
   # labelbottom="off",
    labelleft="on")

    title("t= $tiempo ms")
    imagen=imshow((ejemplo), origin="lower", interpolation="nearest",
    cmap="bwr", 
                      vmin=-limites,vmax=limites, extent=[0.5,64.5,0.5,64.5])

cbx=colorbar(imagen, fraction=0.045, ticks=[]) 

cbx[:set_label]("<-Sink | Source->")


bolitasrojas=0
bolitasazules=0
minlargo=4
marcagor=8
muevemin=0.5
limcsd=80
tight_layout()


for p in values(cadpos)
    gordura=abs.(map(Float32,p[:,3]))
    desviax=std(map(Float32,p[:,1]))
    desviay=std(map(Float32,p[:,2]))
    longus,falsus=size(p)
    if (mean(gordura)> marcagor && (desviax>muevemin || desviay>muevemin) && longus> minlargo )
        #Si son chiquitos no nos interesan.
        xxpos=p[:,1]
        yypos=p[:,2]    
        tiempos=p[:,4]/7022    
        plot(xxpos, yypos, marker="o", markersize=1, color="maroon", lw=0.5)
        #bolitasrojas=scatter(xxpos,yypos, s=gordura, 
        #edgecolors="none", color="r" )
        inicios=scatter(xxpos[1],yypos[1], s=10, edgecolors="black", 
        linewidth=1,
        facecolor="maroon", marker="o" )
        finales=scatter(xxpos[end],yypos[end], s=10, edgecolors="black", 
        linewidth=1,facecolor="maroon", marker="^", label="Final" )
      #  annotate(round(tiempos[1]*1000,1),  (xxpos[1], yypos[1]), fontsize=7, color="black")
       # annotate(round(tiempos[end]*1000,1), (xxpos[end], yypos[end]), fontsize=7, color="green")

    end
end



for p in values(cadneg)
    gordura=abs.(map(Float32,p[:,3]))
    desviax=std(map(Float32,p[:,1]))
    desviay=std(map(Float32, p[:,2]))
    longus,falsus=size(p)
    if mean(gordura)>marcagor && (desviax>muevemin || desviay > muevemin) &&longus> minlargo
        #Si son chiquitos no nos interesan.
        xxpos=p[:,1]
        yypos=p[:,2]    
        tiempos=p[:,4]/7022    
        plot(xxpos, yypos, marker="o", markersize=1, color="midnightblue", lw=0.5)
        #bolitasazules=scatter(xxpos,yypos, s=gordura,
        #edgecolors="none", color="b")
        inicios=scatter(xxpos[1],yypos[1], s=10, edgecolors="black", 
        linewidth=1,
        facecolor="midnightblue", alpha=0.7, marker="o", label="Final" )
        finales=scatter(xxpos[end],yypos[end], s=10, edgecolors="black", 
        linewidth=1,
        facecolor="midnightblue", marker="^", label="Final" )
      #  annotate(round(tiempos[1],4),  (xxpos[1], yypos[1]), fontsize=7, color="black")
      #  annotate(round(tiempos[end],4), (xxpos[end], yypos[end]), fontsize=7, color="green")

    end
end

outname=string(nombregeneral, "-Tray-t-$t-medreqs.png")
savefig(outname, dpi=90)
```
