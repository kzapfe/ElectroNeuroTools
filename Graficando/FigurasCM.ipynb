```julia
function AntesQue(Datos::Array, tiempo)
    Cadena=[0 0 0 0]
    for a in eachindex(Datos[:,4])
        if Datos[a,4]+39<tiempo
            Cadena=vcat(Cadena, Datos[a,:])
    end
end
    Cadena=Cadena[2:end,:]
    return Cadena
end


function DespuesQue(Datos::Array, tiempo)
    Cadena=[0 0 0 0]
    for a in eachindex(Datos[:,4])
        if Datos[a,4]+39>tiempo
            Cadena=vcat(Cadena, Datos[a,:])
    end
end
    Cadena=Cadena[2:end,:]
    return Cadena
end
```
